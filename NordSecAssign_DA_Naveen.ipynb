Once done, I merged all the dataframes together to get one data frame containing all the data.

**Merging all 4 files together on the basis of the ID column by using the lambda function**

```python
df_merged = reduce(lambda left,right:
                    pd.merge(left,right,
                    on = ["id"], how = "outer"), [df1,df2,df3,df4])

print("Full dataset")

df_merged
```

```python
df_merged.info() #to know more information about the data such as its datatypes, null values it contains, etc.

'''
<--- Important Findings ---> 

Columns with null values = 9 columns
redirects = contain nested values should be split
response_code datatype = should be an integer
domain_age_years datatype = should be an integer
'''
```

**Correct the datatype of the columns**

```python
df_merged['response_code'] = pd.to_numeric(df_merged['response_code'], errors='coerce')
df_merged['domain_age_years'] = pd.to_numeric(df_merged['domain_age_years'], errors='coerce')


df_merged = df_merged.astype({'redirects':'string'}) 

df_merged.head()
```

**Work on 'Redirects column**

**First: Removing [ ] brackets from the column data**

```python
df_merged['redirects'] = df_merged['redirects'].str.strip('[]').str.replace("'", "")
df_merged

df_merged.head()
```

**Second: Create two new columns**

```python
#Splitting the redirects column into two different columns

df_merged[['has_redirect', 'url_redirects']] = df_merged['redirects'].str.split(', ', n=1, expand=True)
df_merged
```

**Dropping the original "redirects" column**

```python
df_merged = df_merged.drop(columns=['redirects'])

df_merged
```

# Feature Engineering

Doing feature engineering on two columns named 'reputation_category' & 'true_reputation' as 'malicoius' or 'non_malicious'
Making a dictionary by mapping function to assign values 

```python
rep_category_map = {
    'good': 'NON_MALICIOUS',
    'neutral': 'NON_MALICIOUS',
    'bad': 'MALICIOUS'
}

true_rep_map = {
    'clean': 'NON_MALICIOUS',
    'unknown': 'NON_MALICIOUS',
    'phishing': 'MALICIOUS',
    'scam': 'MALICIOUS',
    'malware': 'MALICIOUS'
}
```

Removing Camel casing and making lowercase and clean before mapping

```python
df_merged['reputation_category'] = df_merged['reputation_category'].str.lower().str.strip()
df_merged['true_reputation'] = df_merged['true_reputation'].str.lower().str.strip()


df_merged['reputation_category_stand'] = df_merged['reputation_category'].map(rep_category_map)
df_merged['true_reputation_stand'] = df_merged['true_reputation'].map(true_rep_map)
```

Creating is_accurate column:

Accurate (1); Inaccurate (0)

```python
df_merged['is_accurate'] = (df_merged['reputation_category_stand'] == df_merged['true_reputation_stand']).astype(int)

df_merged
```

**Dropping Reputation_category & True_reputation column**

```python
df_merged = df_merged.drop(columns=['reputation_category', 'true_reputation'])

df_merged
```
